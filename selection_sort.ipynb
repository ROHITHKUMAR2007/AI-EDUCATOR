```python
#ROHITH KUMAR S
#Selection Sort in Descending Order
numbers = [23, 5, 91, 12, 44, 7, 68, 30]
n = len(numbers)
print("Original list:", numbers)
for i in range(n):
    max_index = i    
    for j in range(i + 1, n):
        if numbers[j] > numbers[max_index]:
            max_index = j
    numbers[i], numbers[max_index] = numbers[max_index], numbers[i]
    print(f"After placing element {i+1}:", numbers)
print("Final sorted list (Descending):", numbers)
```

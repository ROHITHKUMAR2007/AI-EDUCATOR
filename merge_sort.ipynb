```python
#ROHITH KUMAR S
#Merge sort to sort a list of numbers in ascending order
numbers = [42, 7, 19, 3, 25, 11, 8, 30, 2, 15]
print("Original list:", numbers)
size = 1
n = len(numbers)
while size < n:
    print(f" \nMerging sublists of size {size} ")    
    for start in range(0, n, 2 * size):
        mid = min(start + size, n)
        end = min(start + 2 * size, n)        
        left = numbers[start:mid]
        right = numbers[mid:end]       
        i = j = 0
        merged = []      
        # Merge two sublists
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                merged.append(left[i])
                i += 1
            else:
                merged.append(right[j])
                j += 1      
        # Add remaining elements
        merged.extend(left[i:])
        merged.extend(right[j:])        
        print(f"Merging {left} and {right} -> {merged}")        
        # Replace in original list
        numbers[start:end] = merged   
    size *= 2  # Double the sublist size each round
print("\nFinal sorted list (Ascending):", numbers)
```

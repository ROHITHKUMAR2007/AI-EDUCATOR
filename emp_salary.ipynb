```python
#ROHITH KUMAR S
salary = int(input("Enter Base salary:"))
experience = int(input("Enter Experience in years:"))
rating = int(input("Enter Performance rating: "))
if experience >= 10:
    increase = 0.15
elif experience >= 5:
    increase = 0.10
elif experience >= 2:
    increase = 0.05
else:
    increase = 0
salary = salary + (salary * increase)
if rating == 5:
    salary = salary + (salary * 0.10)
elif rating == 4:
    salary = salary + (salary * 0.05)
print("New salary:", int(salary))
```

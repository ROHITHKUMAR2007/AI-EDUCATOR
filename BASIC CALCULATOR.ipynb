```python
#ROHITH KUMAR S
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
add = num1 + num2
sub = num1 - num2
multiply = num1 * num2
division = num1 / num2
floor = num1 // num2
exponent = num1 ** num2
print(f"For numbers {num1} and {num2}: addition = {add}, subtraction = {sub}, multiplication = {multiply}, division = {division}, floor division = {floor}, and exponentiation = {exponent}.")
    
```

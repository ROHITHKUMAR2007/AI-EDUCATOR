```python
#ROHITH KUMAR S
# Insertion Sort in Ascending Order
numbers = [34, 2, 19, 8, 27, 3, 15]
print("Original list:", numbers)
for i in range(1, len(numbers)):
    current_value = numbers[i]
    j = i - 1
    while j >= 0 and numbers[j] > current_value:
        numbers[j + 1] = numbers[j]
        j -= 1
    numbers[j + 1] = current_value
    print(f"After inserting element {i}:", numbers)
    print(f"Sorted portion: {numbers[:i+1]} | Unsorted portion: {numbers[i+1:]}")
print("Final sorted list (Ascending):", numbers)
```

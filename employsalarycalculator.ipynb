```python
#ROHITH KUMAR S

num = int(input("Enter number of employees: "))

for i in range(1, num + 1):
    print(f"\nEmployee {i}")

    base_salary = float(input("Enter base salary: "))
    experience = float(input("Enter experience in years: "))
    rating = int(input("Enter performance rating (1-5): "))

    if base_salary > 0 and experience >= 0:
        if 1 <= rating <= 5:
            
            if experience >= 10:
                salary = base_salary + (base_salary * 0.15)
            elif experience >= 5:
                salary = base_salary + (base_salary * 0.10)
            elif experience >= 2:
                salary = base_salary + (base_salary * 0.05)
            else:
                salary = base_salary

            if rating == 5:
                salary = salary + (salary * 0.10)
            elif rating == 4:
                salary = salary + (salary * 0.05)
            else:
                salary = salary

            print("New base salary:", salary)

        else:
            print("Invalid performance rating! Please enter between 1 and 5.")
    else:
        print("Invalid input! Salary must be positive and experience cannot be negative.")
```
